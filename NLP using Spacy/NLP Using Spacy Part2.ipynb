```python
# Word to find similar words for
word_to_find_similar = "country"

# Check if the word is in the vocabulary
if word_to_find_similar not in nlp.vocab:
    print(f"The word '{word_to_find_similar}' is not in the vocabulary.")
else:
    # Get the word vector for the target word
    target_word_vector = nlp.vocab[word_to_find_similar].vector

    # Find the most similar words
    most_similar_words = []
    for word in nlp.vocab:
        if word.has_vector and word.is_lower:
            similarity = word.vector.dot(target_word_vector)
            most_similar_words.append((word.text, similarity))

    # Sort the list of similar words based on similarity score
    most_similar_words.sort(key=lambda x: -x[1])

    # Print the top 10 most similar words
    print(f"The top 10 most similar words to '{word_to_find_similar}' are:")
    for word, similarity in most_similar_words[:10]:
        print(f"{word}: {similarity:.2f}")
```

```python
doc1 = nlp('I like salty fries and hamburgers.')
doc2 = nlp('Fast food tastes very good.')
```

```python
print(doc1,'<->',doc2,doc1.similarity(doc2))
```
