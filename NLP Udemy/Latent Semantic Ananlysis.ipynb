```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# !wget -nc https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/nlp_class/all_book_titles.txt
```

```python
wordnet_lemmatiazer = WordNetLemmatizer()
```

```python
titles = [line.rstrip()  for line in open('all_book_titles.txt')]
```

```python
stops = set(stopwords.words('english'))
```

```python
stops = stops.union({
    'introduction','edition','series','application','approach','card'
    ,'access','package','plus','etext','brief','vol','fiundamental',
    'guide','essential','printed','third','second','fourth','volume'
})
```

```python
def my_tokenizer(s):
  s = s.lower()

  tokens = nltk.tokenize.word_tokenize(s)

  tokens = [t for t in tokens if len(t) > 2]

  tokens = [wordnet_lemmatiazer.lemmatize(t) for  t in tokens]

  tokens = [t for t in tokens if t not in stops]

  return tokens

```

```python
vectorizer = CountVectorizer(binary=True, tokenizer=my_tokenizer)
```

```python
X = vectorizer.fit_transform(titles)
```

```python
index_word_map = vectorizer.get_feature_names_out()
```

```python
X = X.T
```

```python
svd = TruncatedSVD()
Z = svd.fit_transform(X)
```

```python
import plotly.express as px
```

```python
fig = px.scatter(x=Z[:,0],y=Z[:,1],text=index_word_map,size_max=60)
fig.update_traces(textposition='top center')
fig.show()
```
