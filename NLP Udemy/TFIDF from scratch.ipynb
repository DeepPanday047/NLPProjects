```python
df = pd.read_csv('bbc_text_cls.csv')
df.head()
```

```python
# populate word2idx
# convert documents into sequnces of ints / ids / indices
idx = 0
word2idx = {}
tokenized_docs =[]

for doc in df['text']:
  words = word_tokenize(doc.lower())
  doc_as_int = []

  for word in words:
    if word not in word2idx:
      word2idx[word] = idx
      idx += 1

      # save for later
    doc_as_int.append(word2idx[word])
  tokenized_docs.append(doc_as_int)
```

```python
 # Reverse mapping
 # ifyou do it smarter you can store it as a list

idx2word = {v:k for k, v in word2idx.items()}

```

```python
# number of documents
N = len(df['text'])
N
```

```python
# number of words
V = len(word2idx)
V
```

```python
# instantiate term-frequency matrix
# note: could have also used count vectorizer

tf = np.zeros((N,V))
```

```python
# Populate term-frequency counts
for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    tf[i,j] += 1
```

```python
# compute IDF
document_freq = np.sum(tf > 0, axis=0) # document frequencies
idf = np.log(N / document_freq)
```

```python
# Compute TF -IDF
tf_idf = tf * idf
```

```python
np.random.seed(42)
```

```python
# Pick a random document, show the top 5 terms (in terms of tf_idf score)
i = np.random.choice(N)
row = df.iloc[i]
print('Label:', row['labels'])
print('Text:', row['text'].split('\n', 1)[0])
print('Top 5 terms:')

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
  print(idx2word[j])
```
