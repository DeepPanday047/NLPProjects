```python
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors

# Convert GloVe format to Word2Vec format
glove_input_file = 'glove.6B.300d.txt'
word2vec_output_file = 'glove.6B.300d.word2vec'
glove2word2vec(glove_input_file, word2vec_output_file)

# Load Word2Vec format GloVe vectors using gensim
glove_model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)
```

```python
def find_analogies(w1, w2, w3):
  w1, w2, w3 = w1.lower(), w2.lower(), w3.lower()
  if w1 in glove_model and w2 in glove_model and w3 in glove_model:
    analogy_vector = glove_model[w1] - glove_model[w2] + glove_model[w3]
    closest_words = glove_model.similar_by_vector(analogy_vector, topn=1)
    return closest_words[0][0]
  else:
     return 'Words not in vocabulary'
```

```python
result_analogy = find_analogies('king','man','woman')
print('Analogy:', result_analogy)
```

```python
def nearest_neighbors(word, k=5):
  word = word.lower()
  if word in glove_model:
    closest_words = glove_model.similar_by_word(word,topn=k)
    return [word[0]for word in closest_words]
  else:
    return 'Word not in vocabulary'
```

```python
result_neighbors = nearest_neighbors('king')
print('Nearest Neighbors:',result_neighbors)
```
